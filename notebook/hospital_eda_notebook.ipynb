# Hospital Readmission - Data Cleaning & EDA

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Load dataset
df = pd.read_csv("diabetic_data.csv", na_values=['?'])
df.head()
```

```
# Copy dataset
df_clean = df.copy()

# Drop unnecessary columns
df_clean.drop(columns=['weight', 'payer_code'], inplace=True)

# Fill missing values
df_clean['medical_specialty'].fillna('Unknown', inplace=True)
df_clean['race'].fillna('Unknown', inplace=True)

df_clean.head()
```

```
# Remove duplicate patients
df_clean = df_clean.sort_values('encounter_id') \
                   .drop_duplicates(subset='patient_nbr', keep='first')

# Remove hospice/deceased patients
df_clean = df_clean[~df_clean['discharge_disposition_id']
                    .isin([11,13,14,19,20,21])]

df_clean.shape
```

```
# Create target variable
df_clean['readmitted_binary'] = (df_clean['readmitted'] == '<30').astype(int)

df_clean[['readmitted','readmitted_binary']].head()
```

```
# Simple EDA
df_clean['readmitted_binary'].value_counts()

sns.countplot(x='readmitted_binary', data=df_clean)
plt.title("Readmission Distribution")
plt.show()
```

```
# Save cleaned dataset
df_clean.to_csv("cleaned_data.csv", index=False)

print("Cleaned data saved!")
```
